```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/malladhanraj/Downloads/sales_data.csv')
df.head()
```

```python
# dropping duplicates
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df_1 = df[['product','price']].groupby('product').sum()
df_1.columns = ['total_revenue']
df_1.reset_index(inplace=True)
df_1.head()
```

```python
df_2 = df[['customer_id','price']].groupby('customer_id').sum().sort_values('price', ascending=False)
df_2.columns = ['total_spending']
df_2.reset_index(inplace=True)
df_2[:5]
```

```python
df['month'] = pd.to_datetime(df['order_date']).dt.strftime('%Y-%m')
df.head()
```

```python
df_3 = df['month'].value_counts().reset_index()
df_3.columns = ['month', 'total_orders']
df_3.head()
```

```python
df_1.to_csv('/Users/malladhanraj/Downloads/product_revenue.csv', index=False)
df_2.to_csv('/Users/malladhanraj/Downloads/top_customers.csv', index=False)
df_3.to_csv('/Users/malladhanraj/Downloads/monthly_orders.csv', index=False)
```
